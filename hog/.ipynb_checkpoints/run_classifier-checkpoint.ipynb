```python
from test_face_classifier import test_face_classifier_l
from test_face_classifier import test_face_classifier_rf
from get_testing_data import get_testing_data
import numpy as np
import matplotlib.pyplot as plt
```

```python
# save hog descriptors for testing data as csv
descriptors, classes, races = get_testing_data(300, 9, False)
np.savetxt("hog_descriptors.csv", descriptors, delimiter=",")
np.savetxt("hog_races.csv", races, delimiter=",")
```

```python
# Run classifier over num_trials and return average testing accuracies and false positive rate
# training with normal nonfaces, random forest
def run_classifier(num_trials, classifier):
    avg_overall = np.zeros(5)
    avg_white = np.zeros(5)
    avg_black = np.zeros(5)
    avg_falsepos = np.zeros(5)
    for i in range(num_trials):
        print("trial " + str(i))
        overall_acc = np.zeros(5)
        white_acc = np.zeros(5)
        black_acc = np.zeros(5)
        false_pos = np.zeros(5)
        overall, white, black, falsepos = classifier(300, 300, 9, False, 'train_0', 'training_nonfaces')
        overall_acc[0] = overall
        white_acc[0] = white
        black_acc[0] = black
        false_pos[0] = falsepos
        overall, white, black, falsepos = classifier(300, 300, 9, False, 'train_25', 'training_nonfaces')
        overall_acc[1] = overall
        white_acc[1] = white
        black_acc[1] = black
        false_pos[1] = falsepos
        overall, white, black, falsepos = classifier(300, 300, 9, False, 'train_50', 'training_nonfaces')
        overall_acc[2] = overall
        white_acc[2] = white
        black_acc[2] = black
        false_pos[2] = falsepos
        overall, white, black, falsepos = classifier(300, 300, 9, False, 'train_75', 'training_nonfaces')
        overall_acc[3] = overall
        white_acc[3] = white
        black_acc[3] = black
        false_pos[3] = falsepos
        overall, white, black, falsepos = classifier(300, 300, 9, False, 'train_100', 'training_nonfaces')
        overall_acc[4] = overall
        white_acc[4] = white
        black_acc[4] = black
        false_pos[4] = falsepos
        # sum 
        avg_overall = np.add(avg_overall, overall_acc)
        avg_white = np.add(avg_white, white_acc)
        avg_black = np.add(avg_black, black_acc)
        avg_falsepos = np.add(avg_falsepos, false_pos)
    # average
    avg_overall = avg_overall / num_trials
    avg_white = avg_white / num_trials
    avg_black = avg_black / num_trials
    avg_falsepos = avg_falsepos / num_trials
    
    return (avg_overall, avg_white, avg_black, avg_falsepos)
```

## Logistic

```python
# run logistic classifier with 10 trials
avg_overall_l, avg_white_l, avg_black_l, avg_falsepos_l = run_classifier(10, test_face_classifier_l)
```

```python
# plot accuracies
x = [0, 25, 50, 75, 100]
plt.plot(x, avg_overall_l, label='overall')
plt.plot(x, avg_white_l, label='light-skinned')
plt.plot(x, avg_black_l, label='dark-skinned')
plt.legend(loc="lower right")
plt.title("Average testing accuracies with HOG, logistic function")
plt.xlabel("Percentage of dark-skinned faces in training dataset")
plt.ylabel("Average accuracy")
plt.show()
```

```python
# plot average false positive rates
x = [0, 25, 50, 75, 100]
plt.plot(x, avg_falsepos_l)
plt.title("Average false positive rates with HOG, logistic function")
plt.xlabel("Percentage of dark-skinned faces in training dataset")
plt.ylabel("Average false positive rate")
plt.show()
```
