```python
import cv2
import matplotlib.pyplot as plt
```

```python
face1 = cv2.imread('false_pos/wrong_7.jpg', cv2.IMREAD_GRAYSCALE)
face2 = cv2.imread('false_pos/wrong_8.jpg', cv2.IMREAD_GRAYSCALE)
face3 = cv2.imread('false_pos/wrong_0.jpg', cv2.IMREAD_GRAYSCALE)
face4 = cv2.imread('false_pos/wrong_4.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(face1, cmap='gray')
axarr[0,1].imshow(face2, cmap='gray')
axarr[1,0].imshow(face3, cmap='gray')
axarr[1,1].imshow(face4, cmap='gray')
f.suptitle('HOG false positives', fontsize=12)
f.savefig('hog_false_pos.png', dpi=300)
```

```python
face1 = cv2.imread('false_neg/hard_3.jpg', cv2.IMREAD_GRAYSCALE)
face2 = cv2.imread('false_neg/hard_1.jpg', cv2.IMREAD_GRAYSCALE)
face3 = cv2.imread('false_neg/hard_16.jpg', cv2.IMREAD_GRAYSCALE)
face4 = cv2.imread('false_neg/hard_4.jpg', cv2.IMREAD_GRAYSCALE)
```
